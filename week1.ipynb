# Hello World of Neural Networks

## Importing necessary libraries

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

## Define and Compile Neural Network

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

Compile the model with Stochastic Gradient Descent and use MSE as the loss function

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

## Data
Data generated by y = 2x+1

```python
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

## Training

```python
model.fit(xs, ys, epochs=500, verbose=False)
```

## Evaluation

```python
print(model.predict([10]))
```
